## 회귀(Regression)

회귀 (Regression)
- 분류와 더불어 지도학습의 가장 큰 축 중 하나
- 사실 분류보다 회귀가 현업 업무 분석가들이나 데이터 분석가들이 많이 활용
- 왜냐하면 대부분 기업에서 사용하는 예측은 숫자적 예측이 많기 때문에 회귀 사용(ex. 다음 달 매출액이 얼마가 될 것인가?)
- 회기가 지도학습에서 가장 많이 활용되는 분야

- 현대 통계학을 이루는 큰 축이고, 거의 모든 분야에서 회귀가 많이 활용되고 있음
- 우리는 머신러닝 관점에서의 회귀 사용


### 머신러닝의 회귀 개요

회귀 
- 여러 개의 독립변수와 한 개의 종속변수 간의 상관관계를 모델링하는 기법
- Y = W1*X1 + W2*X2 + W3*X3 + ... + Wn*Xn
- 독립 변수 (X) : 서로 간 상관관계가 적고 독립립적인 변수.
    - 피처.(아파트 크기, 방 개수, 주변 지하철역 수, 주변 학군)
- 종속 변수  (Y): 독립 변수의 영향을 받는 변수. 
    - 결정값. (아파트값)
- 회귀 계수 (W) : 독립 변수의 값에 영향을 미치는 것


1. 최적의 회귀계수를 찾는 것
    - 주어진 피처와 결정값 데이터로 학습
2. 최적의 회귀계수를 찾아 회귀식을 만들어서 새로 들어오는 데이터에 대해 회귀식을 적용시켜 예측을 진행

회귀 유형
- 선형/비선형 : 회귀 계수의 선형/비선형 여부에 따라
- 단일 회귀 / 다중 회귀 : 독립변수 개수에 따라

분류와 회귀 차이점 : 결과값의 차이

분류 : 카테고리 값 반환 (이산값)
- 0/1, 이미지1/이미지2, 카테고리값 0, 1, 2, negative/positive
- 사자, 고양이, 강아지

회기 : 숫자값 (연속값)
- 아파트 값 예측
- 매출 증가액 예측

**선형 회귀**
- 가장 많이 사용되는 회귀
- 실제 값과 예측값의 차이(오류의 제곱 값)을 최소화하는 직선형 회귀선을 최적화하는 방식
- 실제값과 예측값의 차이인 오류를 최소로 줄일 수 있는 선형 함수를 찾는 것이 선형회귀
    - 선형 함수에 독립변수(피처)를 입력해서 종속변수(타깃값, 예측값)을 예측하는 것
- 규제 방법에 따라 여러 유형으로 나뉨
    - 규제 : 일반적인 선형 회기의 과적합 문제를 해결하기 위해서 회기 계수에 패널티 값을 적용하는 것


**대표적인 선형 회귀 모델**
- 일반 선형 회귀 : 규제를 적용하지 않은 모델로 예측값과 실제값의 손실함수(RSS)를 최소화할 수 있도록 최적화 
    - 손실함수: 출력값과 정답의 오차를 정의하는 함수

- 릿지 (Ridge) : 선형 회귀에  L2규제를 추가한 회귀 모델
- 라쏘 (Lasso) : 선형 회귀에  L1규제를 적용한 모델
- 엘라스틱넷 (ElasticNet) : L2, L1 규제를 함께 결합한 모델
- 로지스틱 회귀 (Logistic Regression) : 분류에서 사용되는 선형 모델


**오류의 합 구하는 방식 (손실함수)**
- 오류 값은 (-오류 +오류)가 있어서 그냥 더하면 상쇄되어 원치 않는 결과가 발생함
- 보통 오류 합을 계산할 때는 절대값을 취하거나 **오류 값의 제곱을 구해서 더하는 방식(RSS)** 사용
- 학습을 통해 이 RSS를 최소로 하는 회귀 계수를 찾는 것 머신러닝 기반 회귀의 핵심 사항

RSS (Residual Sum of Square) = Error의 제곱


- 가속도가 계속 증가하면서 속도 증가
- 더 이상 가속도가 증가하지 않으면 그때가 최고 속도
- 이후 가속도가 마이너스가 되면 속도 떨어짐
- 가속도는 속도의 미분으로 구함
- 속도(2차함수) 미분해서 1차함수의 기울기(가속도)를 구함

```python
# 데이터 생성
np.random.seed(0)

# X값 생성(X축 값)
X = 2*np.random.rand(100,1)

# y축 값 생성 : y= 4X1 + 6 (w1=4, w0=6)
# 노이즈 값을 더해서 생성
y = 6 + 4*X + np.random.randn(100,1)
```

```python
# X, y 데이터 셋 scatter plot으로 시각화
plt.scatter(X,y)
```

#### 단순 선형회귀
- LinearRegression()
- fit(X,y) 학습
- predict(test 값) 예측

```python
from sklearn.linear_model import LinearRegression
line_fitter = LinearRegression() # 객체 생성
line_fitter.fit(X, y) # 학습
```

```python
# X가 1.5248의 새로운 데이터의 y값을 예측
y_predicted = line_fitter.predict(np.array([[1.5248]]))
y_predicted
```

```python
# 회귀계수 : coef_  속성
line_fitter.coef_
```

## 채워넣기

```python
# 단순 선형회귀 시각화
plt.plot(X, y, 'o')
plt.plot(X,line_fitter.predict(X))
plt.show()
```

## 사이킷런 LinearRegression 클래스




**사이킷런의 linear_model 모듈은 매우 다양한 종류의 선형 기반 회귀를 클래스로 구현해서 제공**

LinearRegression 클래스
- 예측값과 실제값의 RSS를 최소화해서 OLS(Ordinary Least Squares) 추정 방식으로 구현한 클래스
- fit() 메서드로서 X, y 배열을 입력 받으면
- 회귀 계수(Coefficients)인 W를 coef_ 속성에 저장
- intercept_ : intercept 값 (절편)


선형 회귀의 다중 공산성 문제
- 일반적으로 선형 회귀는 입력 피처의 독립성에 많은 영향을 받음
- 피처간의 상관관계가 매우 높은 경우 분산이 매우 커져서 오류에 매우 민감
- 일반적으로 상관관계가 높은 피처가 많은 경우 
- 독립적인 중요한 피처만 남기고 제거하거나 규제 적용

다중공선성 
- 회귀분석에서 독립변수들 간에 강한 상관관계가 나타나는 문제

상관관계 높은 피처들
- 다른 피처에서 파생된 피처


회귀 평가 지표

MSE (Mean Squared Error) : 실제값과 예측값의 차이를 제곱해서 평균한 것  
    
RMSE (Root Mean Squared Error) : 
- MSE 값은 오류의 제곱을 구하므로 실제 오류 평균보다 더 커지는 특성이 있으므로 MSE에 루트를 쒸운 것
- 사이킷런에서는 제공하지 않기 때문에 MSE에 제곱근을 씌워서 계산하는 함수 직접 만들어서 사용
- 값이 작을수록 성능이 좋음

R제곱 : 분산 기반으로 예측 성능 평가. 
- 실제값의 분산 대비 예측값의 분산 비율을 지표로 하며
- 1에 가까울수록 예측 정확도가 높음
